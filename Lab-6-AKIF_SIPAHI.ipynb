```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])  

y_or = np.array([0, 1, 1, 1])   

perceptron_and = Perceptron()
perceptron_and.fit(X, y_and)
y_pred_and = perceptron_and.predict(X)
and_accuracy = accuracy_score(y_and, y_pred_and)
print(f'AND Function Accuracy: {and_accuracy * 100:.2f}%')

perceptron_or = Perceptron()
perceptron_or.fit(X, y_or)
y_pred_or = perceptron_or.predict(X)
or_accuracy = accuracy_score(y_or, y_pred_or)
print(f'OR Function Accuracy: {or_accuracy * 100:.2f}%')

def plot_decision_boundary(perceptron, X, y, title):
    xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 100), np.linspace(-0.5, 1.5, 100))
    Z = perceptron.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
    plt.title(title)
    plt.xlabel('Input 1')
    plt.ylabel('Input 2')
    plt.show()

plot_decision_boundary(perceptron_and, X, y_and, "Perceptron Decision Boundary for AND")
plot_decision_boundary(perceptron_or, X, y_or, "Perceptron Decision Boundary for OR")
```
